```python
import pandas as pd
from matplotlib import pyplot as plt
```

# Accuracy test analysis

```python
result1_df = pd.read_csv("test1_result.csv")
result1_df
```

```python
result2_df = pd.read_csv("test2_result_trial2.csv")
result2_df
```

```python
#compare sentiment distributios for both tests
result1_df['Neu'].plot(kind='hist', bins=20, title='Neu')
result2_df['Neu'].plot(kind='hist', bins=20, title='Neu')
# result1_df['Neg'].plot(kind='hist', bins=20, title='Neg')
# result2_df['Neg'].plot(kind='hist', bins=20, title='Neg')
# result1_df['Pos'].plot(kind='hist', bins=20, title='Pos')
# result2_df['Pos'].plot(kind='hist', bins=20, title='Pos')
plt.gca().spines[['top', 'right',]].set_visible(False)

```

```python
plt.hist(result1_df['Neu'],
         alpha=0.5, # the transaparency parameter
         label='Neu1',
         color="blue",edgecolor='black', linewidth=1.2)

plt.hist(result2_df['Neu'],
         alpha=0.2,
         label='Neu2',
         color="orange",edgecolor='black', linewidth=1.2)

plt.legend(loc='upper right')
plt.title('Overlapping with both alpha=0.5')
plt.show()
```

```python

avg_compare = {
    "Sentiments" : ['Pos', 'Neg', 'Neu'],
    "Before_leet_translate": [result1_df['Pos'].mean(), result1_df['Neu'].mean(), result1_df['Neg'].mean()],
    "After_leet_translate": [result2_df['Pos'].mean(), result2_df['Neu'].mean(), result2_df['Neg'].mean()],
}
avg_compare_df = pd.DataFrame(avg_compare)
avg_compare_df
```

```python
meanPos = pd.DataFrame(avg_compare)
meanPos.index = meanPos["Sentiments"]
plt.figure().set_figwidth(15)
plt.figure().set_figheight(50)
meanPos.plot.bar(title="Positive Sentiment Analysis by keywords")
plt.ylabel("Sentiment Score")

plt.show()
```
