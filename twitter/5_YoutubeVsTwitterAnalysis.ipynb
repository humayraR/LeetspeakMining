```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# leet_df.to_csv("twitter_leet_df.csv")
yt_avg_df = pd.read_csv("youtube_sentiment_analysis.csv")
avg_compare_df = pd.read_csv("avg_compare_tweet.csv")
```

```python
avg_compare_df.set_index("Sentiments")
tw_yt_avg_join = avg_compare_df.set_index('Sentiments').join(yt_avg_df.set_index('Sentiments'), on='Sentiments')
tw_yt_avg_join
```

```python
# tw_yt_avg_join.index = tw_yt_avg_join["Sentiments"]
plt.figure().set_figwidth(15)
plt.figure().set_figheight(50)
tw_yt_avg_join.plot.bar(title="Twitter Vs Youtube Overall Sentiment Analysis")
plt.ylabel("Sentiment Score")
plt.legend(loc='lower right')

plt.show()
```
