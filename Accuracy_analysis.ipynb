```python
import pandas as pd
```

# Accuracy test analysis

```python
result_df = pd.read_csv("pyleet_accuracy_result.csv")
result_df
```

```python
yes = result_df[result_df['correct'] =='yes']
#no - try to match words that are already in the dictionary, if can't match, do direct substitution, mostly words  that resemble(close to) some words but not in dictionary
#none - word lengths>=15
yes
```

```python
no = result_df[result_df['correct'] =='no']
no
```

```python
# prompt: Using dataframe result_df: find rows with result_df['correct'] = None

null = result_df[result_df['correct'].isnull()]
null
```

```python
# prompt: Using dataframe result_df: replace all None values with "None" string

result_df = result_df.fillna('None')
```

```python
result_df
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
result_df.groupby('correct').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
plt.xlabel('Word count')
plt.ylabel('Correctness of spelling')
```

```python
# prompt: Using dataframe result_df: suggest plots

result_df.groupby('correct').size().plot(kind='bar')
```

```python
result_df
```

```python
yes.to_csv('yes.csv', index = False)
no.to_csv('no.csv', index = False)
null.to_csv('null.csv', index = False)
```

# Profanity Test analysis

```python
profanity_df = pd.read_csv("profanity_test_result.csv")
profanity_df
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
profanity_df.groupby('correct').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
plt.xlabel('Word count')
plt.ylabel('Correctness of spelling')
```

```python
# prompt: Using dataframe profanity_df: drop unnamed column

profanity_df.drop(columns='Unnamed: 0', axis=1, inplace=True)
```

```python
profanity_df
```

```python
# prompt: Using dataframe profanity_df: get rows with profanity_df['correct'] containing 'yes'

profanity_yes = profanity_df[profanity_df.correct.str.contains('yes')]
profanity_no = profanity_df[profanity_df.correct.str.contains('no')]
profanity_yes
```

```python
profanity_no
```

```python
# prompt: Using dataframe profanity_df: get rows with profanity_df['correct'] containing no 'yes' and no 'no'

profanity_null = profanity_df[~profanity_df['correct'].str.contains('yes|no')]
profanity_null
```

```python
profanity_yes.to_csv('profanity_yes.csv', index = False)
profanity_no.to_csv('profanity_no.csv', index = False)
profanity_null.to_csv('profanity_null.csv', index = False)
```
