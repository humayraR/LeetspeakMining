```python
# prompt: Using dataframe result1_df: find average result1_df['Neg'] values

print(result1_df['Neg'].mean())
print(result2_df['Neg'].mean())
```

```python
avg_compare = {
    "Sentiments" : ['Pos', 'Neg', 'Neu'],
    "Before_leet_translate": [result1_df['Pos'].mean(), result1_df['Neu'].mean(), result1_df['Neg'].mean()],
    "After_leet_translate": [result2_df['Pos'].mean(), result2_df['Neu'].mean(), result2_df['Neg'].mean()],
}
avg_compare_df = pd.DataFrame(avg_compare)
avg_compare_df#.to_csv("youtube_sentiment_analysis.csv", index = False)
```

```python
meanPos = pd.DataFrame(avg_compare)
meanPos.index = meanPos["Sentiments"]
plt.figure().set_figwidth(15)
plt.figure().set_figheight(50)
meanPos.plot.bar(title="Youtube overall Sentiment Analysis")
plt.ylabel("Sentiment Score")

plt.show()
```
